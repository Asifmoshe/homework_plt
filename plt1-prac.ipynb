```python
df = pd.read_csv("C:/Users/SHIREL/Downloads/tips (1).csv")
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 7))


axes[0,0].scatter(df['price_per_person'], df['tip'], color='g')
axes[0,0].set_title('Tip VS Price per person')
axes[0,0].set_xlabel('Price per person')
axes[0,0].set_ylabel('Tip')

max_bill_per_day = df.groupby('day')['total_bill'].max()
axes[0,1].bar(max_bill_per_day.index, max_bill_per_day.values, color='b')
axes[0,1].set_title('Max bill per day')
axes[0,1].set_xlabel('Day')
axes[0,1].set_ylabel('Max bill')

axes[1,0].hist(df['tip'], bins=10, color='r', alpha=0.7)
axes[1,0].set_title('Tips')
axes[1,0].set_xlabel('Tips')
axes[1,0].set_ylabel('N of tip in the table')

df['tip_perc'] = 100 * (df['tip'] / df['total_bill'])
axes[1,1].hist(df['tip_perc'], bins=20, color='y', alpha=0.7)
axes[1,1].set_title('Tip perc')
axes[1,1].set_xlabel('Tip perc')
axes[1,1].set_ylabel('N of tip perc in the table')

plt.tight_layout()
plt.show()
plt.savefig('tips_overview.png')
```
